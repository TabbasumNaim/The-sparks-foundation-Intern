K-Means Clustering





Step-01

Importing The Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

Step-02


Loading The Dataset And Print in The DataFrame

```python
iris = datasets.load_iris()
print("The Given Data Imported")
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head()
```

Step-03

Finding The Optimum Number Of Clusters for K-means

```python
x = iris_df.iloc[:, [0, 1, 2, 3]].values

from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
```

Step-04

Plotting The Results On a Line Graph

```python
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
print("Results Plotted Successfully")
```

Step-05

Applying K-means to The Dataset

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
print("Applied Successfully")
```

Step-06

Visualising The Clusters And Plotting The Centroids Of The Clusters

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()

print("Plotted Successfully")
```
